```python
#Plot the y column in the training dataset in 'line', 'hist' and 'box' plots
plt.style.use('ggplot')
pd.set_option('display.max_rows', 5000)
kinds = ['line', 'hist', 'box']
fig, ax = plt.subplots(len(kinds))
fig.subplots_adjust(left=0, right=2, bottom=0.1, top = 3)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for i, kind in enumerate(kinds):
    ax[i].set_title('y column ' + kind + ' plot of train')
    df_train['y'].plot(ax=ax[i], kind=kind)
```

```python
df_train[df_train['y']>135]['y'].value_counts().sum()
```

```python
# Plot the columns of 'X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8' in bar plot
plt.style.use('ggplot')
pd.set_option('display.max_rows', 5000)
#plt.set_option
columns = ['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']
fig, ax = plt.subplots(len(columns), 2)
fig.subplots_adjust(left=0, right=2, bottom=0.1, top = 10)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
for i, col in enumerate(columns):
    ax[i,0].set_title(col + ' column value counts of train')
    df_train[col].value_counts().plot(ax=ax[i, 0], kind='bar')
    ax[i,1].set_title(col + ' column value counts of test')
    df_test[col].value_counts().plot(ax=ax[i, 1], kind='bar')
```
